```python
import pandas as pd

# Load data into pandas DataFrames
housing_data = ["median_sales_price_data.csv", "median_house_prices.csv"]
dfs = [pd.read_csv(f) for f in housing_data]

# Format the date to be 'YY/MM/DD'
for df in dfs:
    df['DATE'] = pd.to_datetime(df['DATE']).dt.strftime('%Y-%m-%d')


```

```python

# Extract year and month from DATE column and format as 'YYYY-MM'
for df in dfs[1:]:
    df['month'] = pd.to_datetime(df['DATE']).dt.to_period('M').astype(str)

# Select only the year and month from the DATE column of dfs[1] and dfs[2] and format as 'YYYY-MM'
dfs[0]['month'] = pd.to_datetime(dfs[0]['DATE']).dt.to_period('M').astype(str)
dfs[1]['month'] = pd.to_datetime(dfs[1]['DATE']).dt.to_period('M').astype(str)

# Drop the DATE column from dfs[1] and dfs[2]
for df in dfs[1:]:
    df.drop(columns=['DATE'], inplace=True)

# Rename the column in dfs[1] and dfs[2] to 'month'
for df in dfs[1:]:
    df.rename(columns={'month': 'month'}, inplace=True)
```
