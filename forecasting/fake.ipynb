```python
import pandas as pd
import numpy as np
from faker import Faker
import datetime

# Initialize Faker
fake = Faker()

# Generate date range from March 2014 to March 2024 on a weekly basis
start_date = datetime.date(2014, 3, 1)
end_date = datetime.date(2024, 3, 1)
date_range = pd.date_range(start=start_date, end=end_date, freq='W')

# Initialize prices
start_house_price = 374320
end_house_price = 1580455
start_sales_price = 384320
end_sales_price = 1480455

# Calculate the number of weeks in the date range
num_weeks = len(date_range)

# Generate price trends using random walk
np.random.seed(0)  # For reproducibility
house_price = np.linspace(start_house_price, end_house_price, num_weeks)
sales_price = np.linspace(start_sales_price, end_sales_price, num_weeks)

# Add some realistic noise
house_price += np.random.normal(0, 15000, num_weeks)  # 15000 is the noise level
sales_price += np.random.normal(0, 15000, num_weeks)

# Create the DataFrame
data = {
    'Date': date_range.strftime('%Y-%m-%d'),
    'Median House Price': np.round(house_price, 2),
    'Median Sales Price': np.round(sales_price, 2)
}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file in the current directory
csv_filename = 'XX_House_Trend_Data.csv'  # Removed the directory path
df.to_csv(csv_filename, index=False)

csv_filename
```
